# **INTRODUCTION**


Lãng phí chi tiêu quảng cáo là vấn đề lớn mà tất cả các doanh nghiệp phải đối mặt. Chỉ riêng trong năm 2021, 3 tỷ đô la Mỹ chi tiêu cho quảng cáo đã bị lãng phí. Một trong những nguyên nhân chính dẫn đến sự lãng phí này là sai lầm khi nhắm mục tiêu sai đối tượng. Mục tiêu của chiến dịch quảng cáo là thu hút sự chú ý của càng nhiều khách hàng càng tốt. Do đó, nó nên được tiếp thị hướng tới nhóm có nhiều khả năng mua sản phẩm nhất. Đổi lại, mang lại nhiều doanh thu hơn.

**Mục tiêu dự án**
- Trong dự án này, chúng ta sẽ khám phá dữ liệu được thu thập bởi một siêu thị về khách hàng của siêu thị đó. Chúng ta sẽ có một số thống kê về khách hàng nói chung của siêu thị và thói quen của họ. Sau đó, chúng tôi sẽ cố gắng chia khách hàng thành các phân khúc bằng các kỹ thuật phân cụm khác nhau. Chúng tôi sẽ cố gắng đánh giá việc phân cụm bằng một số số liệu phổ biến. Tuy nhiên, chúng ta sẽ phải nghiên cứu sâu về cụm để đánh giá tính đúng đắn của thuật toán phân cụm.

- Vào cuối dự án, chúng ta sẽ có một mô tả thống kê về các phân khúc khách hàng khác nhau và thói quen chi tiêu của họ. Sau đó, siêu thị có thể tận dụng các phân khúc đó để tiếp thị sản phẩm của họ tốt hơn.

**Mô tả nhiệm vụ**
- **Unsupervised Learning Task** - Phân khúc tính cách khách hàng: Bộ dữ liệu này cũng cung cấp cho chúng tôi một số thông tin về khách hàng (Bao gồm thông tin cơ bản và sở thích mua hàng của họ). Vì vậy, chúng tôi có thể thực hiện Phân tích tính cách khách hàng để giúp tìm ra những khách hàng lý tưởng của công ty. Nó giúp doanh nghiệp hiểu rõ hơn về khách hàng của mình và giúp họ lập kế hoạch tiếp thị phù hợp dễ dàng hơn, chẳng hạn như sửa đổi và quảng cáo sản phẩm cho các khách hàng khác nhau theo nhu cầu, hành vi và mối quan tâm cụ thể.

```python
def display_missing(df, feature_cols):
    n_rows = df.shape[0]
    for col in feature_cols:
        missing_count = df[col].isnull().sum()
        if missing_count > 0:
            print(f"{col} has {missing_count*100/n_rows:.2f}% missing values.")
            
display_missing(data, feature_cols)
```

```python
data = data.dropna()
print("The total number of data-points after removing the rows with missing values are:", len(data))
```

Trong bước tiếp theo, ta sẽ tạo một tính năng từ **"Dt_Customer"** cho biết số ngày một khách hàng được đăng ký trong cơ sở dữ liệu của công ty. Tuy nhiên, để đơn giản ta lấy giá trị này so với khách hàng gần đây nhất trong hồ sơ.

Vì vậy, tạo một tính năng **("Customer_For")** của số ngày khách hàng bắt đầu mua sắm tại cửa hàng so với ngày được ghi lại cuối cùng

```python
data["Dt_Customer"] = pd.to_datetime(data["Dt_Customer"])
dates = []
for i in data["Dt_Customer"]:
    i = i.date()
    dates.append(i)  
```

```python
#Created a feature "Customer_For"
days = []
d1 = max(dates)
for i in dates:
    delta = (d1 - i).days
    days.append(delta)
data["Customer_For"] = days
data["Customer_For"] = pd.to_numeric(data["Customer_For"], errors="coerce")
```

Bây giờ chúng ta sẽ khám phá các giá trị duy nhất trong các tính năng phân loại để có ý tưởng rõ ràng về dữ liệu.

```python
print("Total categories in the feature Marital_Status:\n", data["Marital_Status"].value_counts(), "\n")
print("Total categories in the feature Education:\n", data["Education"].value_counts())
```

**Trong phần tiếp theo, ta sẽ thực hiện các bước sau để thiết kế một số tính năng mới:**

* Tạo thuộc tính **"Age"** của khách hàng theo **"Year_Birth"** cho biết năm sinh của người tương ứng tính đến thời điểm gần nhất của năm bộ dữ liệu được thu thập.
* Tạo một thuộc tính  khác **"Spent"** cho biết tổng số tiền mà khách hàng đã chi tiêu trong các danh mục khác nhau trong khoảng thời gian hai năm.
* Tạo thêm một thuộc tính **"Living_With"** ngoài **"Marital_Status"** để trích xuất hoàn cảnh chung sống của các cặp vợ chồng.
* Tạo một thuộc tính  **"Children"** để cho biết tổng số trẻ em trong một hộ gia đình là trẻ em và thanh thiếu niên.
* Để hiểu rõ hơn về hộ gia đình, Tạo thuộc tính cho biết **"Family_Size"**
* Tạo thuộc tính **"Is_Parent"** để hiển thị trạng thái làm cha mẹ
* Cuối cùng, tôi sẽ tạo ba danh mục trong **"Education"** bằng cách đơn giản hóa số lượng giá trị của nó.
* Bỏ một số tính năng dư thừa

```python
#Feature Engineering
#Age of customer today 
data["Age"] = 2015-data["Year_Birth"]

#Total spendings on various items
data["Spent"] = data["MntWines"] + data["MntFruits"] + data["MntMeatProducts"] + data["MntFishProducts"] + data["MntSweetProducts"] + data["MntGoldProds"]

#Deriving living situation by marital status"Alone"
data["Marital_Status"]=data["Marital_Status"].replace({"Married":"Partner", "Together":"Partner", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone",})

#Feature indicating total children living in the household
data["Children"]=data["Kidhome"]+data["Teenhome"]

#Feature for total members in the householde
data["Family_Size"] =  data["Marital_Status"].replace({"Alone": 1, "Partner":2}) + data["Children"]

#Feature pertaining parenthood
data["Is_Parent"] = np.where(data.Children> 0, 1, 0)

#Segmenting education levels in three groups
data["Education"]=data["Education"].replace({"Basic":"Undergraduate","2n Cycle":"Undergraduate", "Graduation":"Graduate", "Master":"Postgraduate", "PhD":"Postgraduate"})

#For clarity
data=data.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold"})

# Clear description of feature
data.rename(columns={'MntWines':'Wines', 'MntFruits':'Fruits', 'MntMeatProducts':'Meats', 'MntFishProducts':'Fish', 'MntSweetProducts':'Sweets', 'MntGoldProds':'Golds'}, inplace=True)
data.rename(columns={'NumWebPurchases':'Web', 'NumCatalogPurchases':'Catalog', 'NumStorePurchases':'Store'}, inplace=True)

#Create 'Num_Accepted' feature from the sum of previous marketting campaigns that were accepted by the customer
data['Num_Accepted'] = data.AcceptedCmp1.values + data.AcceptedCmp2.values + \
                                data.AcceptedCmp3.values + data.AcceptedCmp4.values + data.AcceptedCmp5.values

#Dropping some of the redundant features
to_drop = ["Z_CostContact", "Z_Revenue", "Year_Birth", "ID","Kidhome","Teenhome","Children"]
data = data.drop(to_drop, axis=1)
print("Bây giờ chúng ta có một số thuộc tính mới, hãy xem số liệu thống kê của dữ liệu.")
data.head()
```

# **Data preprocessing & Dimension Reduction**



# Data preprocessing
<a id="4"></a>
 <p style="background-color:#682F2F;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;"></p>

In this section, I will be preprocessing the data to perform clustering operations.

**The following steps are applied to preprocess the data:**

* Label encoding the categorical features
* Scaling the features using the standard scaler 
* Creating a subset dataframe for dimensionality reduction](https://)

```python
#Get list of categorical variables
s = (data.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables in the dataset:", object_cols)
```

```python
#Label Encoding the object dtypes.
LE=LabelEncoder()
for i in object_cols:
    data[i]=data[[i]].apply(LE.fit_transform)
    
print("All features are now numerical")
```

```python
#Creating a copy of data
ds = data.copy()

# creating a subset of dataframe by dropping the features on deals accepted and promotions
cols_del = ["AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3", "AcceptedCmp4", "AcceptedCmp5", "Complain", "Num_Accepted", "Response", "Dt_Customer"]
ds = ds.drop(cols_del, axis=1)

#Scaling
scaler = StandardScaler()
scaler.fit(ds)
scaled_ds = pd.DataFrame(scaler.transform(ds),columns= ds.columns )
print("All features are now scaled")
```

```python
#Scaled data to be used for reducing the dimensionality
print("Dataframe to be used for further modelling:")
scaled_ds.head()
```

<a id="5"></a>
# <p style="background-color:#682F2F;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">DIMENSIONALITY REDUCTION</p>
In this problem, there are many factors on the basis of which the final classification will be done. These factors are basically attributes or features. The higher the number of features, the harder it is to work with it. Many of these features are correlated, and hence redundant. This is why I will be performing dimensionality reduction on the selected features before putting them through a classifier.  
*Dimensionality reduction is the process of reducing the number of random variables under consideration, by obtaining a set of principal variables.* 

**Principal component analysis (PCA)** is a technique for reducing the dimensionality of such datasets, increasing interpretability but at the same time minimizing information loss.

**Steps in this section:**
* Dimensionality reduction with PCA
* Plotting the reduced dataframe

**Dimensionality reduction with PCA**

For this project, I will be reducing the dimensions to 3.

```python
#Initiating PCA to reduce dimentions aka features to 3
pca = PCA(n_components=3)
pca.fit(scaled_ds)
PCA_ds = pd.DataFrame(pca.transform(scaled_ds), columns=(["col1","col2", "col3"]))
PCA_ds.describe().T
```

```python
#A 3D Projection Of Data In The Reduced Dimension
x =PCA_ds["col1"]
y =PCA_ds["col2"]
z =PCA_ds["col3"]
#To plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, c="maroon", marker="o" )
ax.set_title("A 3D Projection Of Data In The Reduced Dimension")
plt.show()
```

# **Unsupervised Task - Customer Personality Segmentation**

<a id="6"></a>
# <p style="background-color:#682F2F;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">CLUSTERING</p>

Now that I have reduced the attributes to three dimensions, I will be performing clustering via Agglomerative clustering. Agglomerative clustering is a hierarchical clustering method.  It involves merging examples until the desired number of clusters is achieved.

**Steps involved in the Clustering**
* Elbow Method to determine the number of clusters to be formed
* Clustering via Agglomerative Clustering
* Examining the clusters formed via scatter plot

```python
# Quick examination of elbow method to find numbers of clusters to make.
print('Elbow Method to determine the number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(PCA_ds)
Elbow_M.show()
```

To examine the clusters formed let's have a look at the 3-D distribution of the clusters. 

We will use k = 4, thus there will be four clusters in our dataset.
We will use KMeans Algorithm for clustering.

```python
km = KMeans(n_clusters = 4,random_state = 42)

y_km = km.fit_predict(PCA_ds)
PCA_ds["Clusters"] = y_km
data["Clusters"]= y_km
```

```python
#Plotting the clusters
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(x, y, z, s=40, c=PCA_ds["Clusters"], marker='o', cmap = 'RdBu_r' )
ax.set_title("The Plot Of The Clusters")
plt.show()
```

```python
y_reduced = km.fit(PCA_ds)
clusters = y_reduced.labels_

# add cluster labels to the original dataset
data['Cluster'] = clusters
data.head()
```
